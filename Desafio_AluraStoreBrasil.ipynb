### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print('Faturamento Total Loja 1:', faturamento_loja1)
print('Faturamento Total Loja 2:', faturamento_loja2)
print('Faturamento Total Loja 3:', faturamento_loja3)
print('Faturamento Total Loja 4:', faturamento_loja4)
```

```python
import matplotlib.pyplot as plt

# Valores de faturamento
faturamentos = [
    faturamento_loja1,
    faturamento_loja2,
    faturamento_loja3,
    faturamento_loja4
]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamentos)

plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento (R$)')
plt.xlabel('Lojas')

# Mostra os valores no topo das barras
for i, valor in enumerate(faturamentos):
    plt.text(i, valor, f"{valor:,.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria

```python
categoria_loja1 = loja.groupby('Categoria do Produto')['Categoria do Produto'].count()
categoria_loja2 = loja2.groupby('Categoria do Produto')['Categoria do Produto'].count()
categoria_loja3 = loja3.groupby('Categoria do Produto')['Categoria do Produto'].count()
categoria_loja4 = loja4.groupby('Categoria do Produto')['Categoria do Produto'].count()

#a quantidade de produtos vendidos por categoria em cada loja
print(categoria_loja1)
print('------------------')
print(categoria_loja2)
print('------------------')
print(categoria_loja3)
print('------------------')
print(categoria_loja4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_categorias = pd.DataFrame({
    'Loja 1': categoria_loja1,
    'Loja 2': categoria_loja2,
    'Loja 3': categoria_loja3,
    'Loja 4': categoria_loja4
}).fillna(0)

# Agrupar categorias
categorias_agrupadas = df_categorias.sum(axis=1)

# Gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(categorias_agrupadas, labels=categorias_agrupadas.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Categorias')
plt.axis('equal')
plt.show()


```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

print('Média de avaliação - Loja 1:', media_loja1)
print('Média de avaliação - Loja 2:', media_loja2)
print('Média de avaliação - Loja 3:', media_loja3)
print('Média de avaliação - Loja 4:', media_loja4)
```

# 4. Produtos Mais e Menos Vendidos

```python
# Loja 1
produtos_loja1 = loja.groupby('Produto').size().sort_values(ascending=False)
print('=== Loja 1 ===')
print('Produtos mais vendidos:')
print(produtos_loja1.head())
print('\nProdutos menos vendidos:')
print(produtos_loja1.tail())

# Loja 2
produtos_loja2 = loja2.groupby('Produto').size().sort_values(ascending=False)
print('\n=== Loja 2 ===')
print('Produtos mais vendidos:')
print(produtos_loja2.head())
print('\nProdutos menos vendidos:')
print(produtos_loja2.tail())

# Loja 3
produtos_loja3 = loja3.groupby('Produto').size().sort_values(ascending=False)
print('\n=== Loja 3 ===')
print('Produtos mais vendidos:')
print(produtos_loja3.head())
print('\nProdutos menos vendidos:')
print(produtos_loja3.tail())

# Loja 4
produtos_loja4 = loja4.groupby('Produto').size().sort_values(ascending=False)
print('\n=== Loja 4 ===')
print('Produtos mais vendidos:')
print(produtos_loja4.head())
print('\nProdutos menos vendidos:')
print(produtos_loja4.tail())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

produtos_todas = (
    pd.concat([loja[['Produto']], loja2[['Produto']], loja3[['Produto']], loja4[['Produto']]])
      .groupby('Produto')
      .size()
      .sort_values(ascending=False)
)

# Selecionar mais e menos vendidos
top10 = produtos_todas.head(10)
bottom10 = produtos_todas.tail(10)

plt.figure(figsize=(14, 7))

# Gráfico dos mais vendidos
plt.subplot(1, 2, 1)
plt.plot(top10.index, top10.values, marker='o', linestyle='-', color='blue')
plt.title('Top 10 Produtos Mais Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(True)

# Gráfico dos menos vendidos
plt.subplot(1, 2, 2)
plt.plot(bottom10.index, bottom10.values, marker='o', linestyle='-', color='orange')
plt.title('Top 10 Produtos Menos Vendidos (Todas as Lojas)')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()

```
